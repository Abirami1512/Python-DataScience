```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("carprices.csv")
df.head()
```

```python
plt.scatter(df['Mileage'],df['Sell Price($)'])
```

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
```

```python
x = df[['Mileage','Age(yrs)']]
y = df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split # the package needed
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train # random values are randomly selected everytime
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
reg.score(x_test,y_test)
```

```python
plt.scatter(df['Mileage'],df['Sell Price($)'])
plt.plot(x_train[['Mileage']],reg.predict(x_train),color ='red')
```

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
plt.plot(x_train[['Age(yrs)']],reg.predict(x_train[['Mileage','Age(yrs)']]),color ='red')
```
