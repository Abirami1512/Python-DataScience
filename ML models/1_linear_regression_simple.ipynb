```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline

plt.xlabel('area')
plt.ylabel('prices')
plt.title('House-prices')
plt.scatter(df.area,df.price) # the function for scatter plot
```

```python
reg = linear_model.LinearRegression() # creating an object for linear reg
#reg.fit(df.area,df.price)----> results in an error
reg.fit(df[['area']],df.price)
# the arguments are x and y vals
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.xlabel('area')
plt.ylabel('prices')
plt.title('House-prices')
plt.scatter(df.area,df.price)
plt.plot(df.area,reg.predict(df[['area']]),color = 'red')
```

```python
df1 = pd.read_csv("canada_per_capita_income.csv")
df1
```

```python
plt.xlabel('year')
plt.ylabel('income')
plt.title('per capita income')
plt.scatter(df1['year'],df1['per capita income (US$)'])
```

```python
reg2 = linear_model.LinearRegression()
reg2.fit(df1[['year']],df1['per capita income (US$)'])
```

```python
reg2.predict([[1976]])
```

```python
plt.xlabel('year')
plt.ylabel('income')
plt.title('per capita income')
plt.scatter(df1['year'],df1['per capita income (US$)'])
plt.plot(df1['year'],reg2.predict(df1[['year']]),color = 'red')
```
