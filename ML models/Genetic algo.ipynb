```python
import pygad
import numpy as np
```

```python
f_ip = [1,2,3,4,5,6]

d_op = 44

def fitness_function(ga_instance,solution,solution_idx):
    op = np.sum(f_ip*solution)
    fitness = 1.0/np.abs(d_op-op)
    return fitness
```

```python
ga = pygad.GA(num_generations = 50,
             num_parents_mating = 4,
             fitness_func = fitness_function,
             sol_per_pop = 8,
             num_genes = len(f_ip),
             mutation_type = None)
```

```python
ga.run()
```

```python
sol,sol_fit,sol_idx = ga.best_solution()

print("solution: ",sol)
print("solution fitness: ",sol_fit)
print("prediction: ",np.sum(sol*f_ip))
```
